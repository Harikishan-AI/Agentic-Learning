```python
from dataclasses import dataclass

@dataclass
class Person:
    name :str
    age: int
    city: str

person = Person(name="John", age=30, city="New York")
print(person)
```

```python
person = Person(name="hari", age=36, city=34)
person
```

```python
##pydantic

from pydantic import BaseModel

class User(BaseModel):
    name: str
    age: int
    is_active: bool

user1 = User(name="John", age=30, is_active=True)
print(user1)
```

```python
user2 = User(name="Jane", age=25, is_active=34)
print(user2)
```

```python

from typing import Optional
from pydantic import BaseModel

class Employee(BaseModel):
    id: int
    name: str
    department: str
    salary: Optional[float] =None
    is_active: Optional[bool] =True
```

```python
e1 = Employee(name="ksihan",id= 1, department="social", salary="45",is_active=False)
e1
```

```python
from typing import List

class classroom(BaseModel):
    room_num: str
    students: List[str]
    capacity: int

c1 = classroom(room_num="so",students=["ha"],capacity=45)
```

```python
c1
```

```python
try:
    classroom(room_num="so",students="hh",capacity=45)
except Exception as e:
    print(e)
```

```python
from pydantic import BaseModel

class Adress(BaseModel):
    street: str
    city: str
    zip_code: str

class Customer(BaseModel):
    name: str
    age: int
    address: Adress
```

```python
c1 = Customer(name="John", age=30, address=Adress(street="123 Main St", city=12, zip_code="12345"))
```

```python
c1
```

```python
from pydantic import BaseModel, Field

class Item(BaseModel):
    name: str=Field(min_length=4)
    price: float=Field(gt=0)
    tax: float=Field(default=0.0)
```

```python
t1 = Item(name="Itefsf", price=1)  
```

```python
t1
```

```python
class User(BaseModel):
    username: str=Field(description="The username of the user", min_length=4)
    age: int=Field(description="The age of the user", ge=18)
    email: str=Field(default_factory=lambda: "user@example.com")

user1 = User( age=18, email="udaasdasdahs@example.com")
print(user1)
```
